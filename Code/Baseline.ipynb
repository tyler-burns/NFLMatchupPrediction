I want to create another baseline that compares the predicted over under line to the actual line to give me a good idea of what's possible.

```python
# Imports

import nflreadpy as nfl
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Set pandas settings
pd.set_option('display.max_columns', None)
```

```python
schedule = nfl.load_schedules(seasons = [2024, 2025])

schedule = schedule.to_pandas()
```

```python
schedule.head(5)
```

```python
schedule = schedule[schedule['total_line'].notna() & schedule['total'].notna()]
```

```python
mse = mean_squared_error(schedule['total'], schedule['total_line'])
mae = mean_absolute_error(schedule['total'], schedule['total_line'])

print(f"MSE: {mse}\nMAE: {mae}\n")
```
